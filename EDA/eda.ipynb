This notebook explores the SkyWest Airlines flight dataset to understand data quality, feature distributions, and patterns that will guide our preprocessing and modeling decisions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

pd.set_option('display.max_columns', None)

# Create plots directory
PLOTS_DIR = 'plots'
os.makedirs(PLOTS_DIR, exist_ok=True)

# Load dataset
df = pd.read_csv('../data/skywest_flights.csv', low_memory=False)
print(f'Shape: {df.shape[0]:,} rows × {df.shape[1]} columns')
```

## 2. Missing Value Analysis

```python
# Missing values — only columns with >0% missing
missing_pct = (df.isnull().sum() / len(df) * 100).round(1)
missing_pct = missing_pct[missing_pct > 0].sort_values(ascending=True)

# Color by severity: orange = partial, red = ≥95%
colors = ['#c44e52' if v >= 95 else '#dd8452' for v in missing_pct.values]

fig, ax = plt.subplots(figsize=(10, 22))
ax.barh(range(len(missing_pct)), missing_pct.values, color=colors)
ax.set_yticks(range(len(missing_pct)))
ax.set_yticklabels(missing_pct.index, fontsize=8)
ax.set_xlabel('Missing (%)')
ax.set_title(f'Missing Data by Feature ({len(missing_pct)} of {df.shape[1]} columns)')

# Legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#dd8452', label='Partial missing'),
                   Patch(facecolor='#c44e52', label='≥95% missing')]
ax.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/missing_data.png', dpi=150)
plt.close(fig)

n_zero = df.shape[1] - len(missing_pct)
n_partial = ((missing_pct > 0) & (missing_pct < 95)).sum()
n_drop = (missing_pct >= 95).sum()
print(f'0% missing (not shown): {n_zero} columns')
print(f'Partial missing (0-95%): {n_partial} columns')
print(f'≥95% missing (to drop): {n_drop} columns')
```

## 3. Target Variable: Arrival Delay

```python
# Drop rows where ArrDelay is missing (cancelled/diverted flights)
print(f'Rows before dropping missing ArrDelay: {len(df):,}')
df_valid = df.dropna(subset=['ArrDelay']).copy()
print(f'Rows after: {len(df_valid):,} (dropped {len(df) - len(df_valid):,})')

# Engineer binary target
df_valid['Delayed'] = (df_valid['ArrDelay'] >= 15).astype(int)

# Class distribution
counts = df_valid['Delayed'].value_counts()
pcts = df_valid['Delayed'].value_counts(normalize=True) * 100

print(f'\nClass Distribution:')
print(f'  On-Time (0): {counts[0]:,} ({pcts[0]:.1f}%)')
print(f'  Delayed (1): {counts[1]:,} ({pcts[1]:.1f}%)')
print(f'  Imbalance ratio: {counts[0]/counts[1]:.1f}:1')
```

```python
# Target class distribution and raw ArrDelay histogram
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Class distribution bar chart
labels = ['On-Time', 'Delayed (≥15 min)']
colors = ['#4c72b0', '#dd8452']
axes[0].bar(labels, [counts[0], counts[1]], color=colors)
axes[0].set_title('Target Class Distribution')
axes[0].set_ylabel('Count')
for i, v in enumerate([counts[0], counts[1]]):
    axes[0].text(i, v + 500, f'{v:,}', ha='center', fontsize=11)

# Raw ArrDelay distribution
axes[1].hist(df_valid['ArrDelay'].clip(-60, 180), bins=80, color='steelblue', edgecolor='white')
axes[1].axvline(x=15, color='red', linestyle='--', linewidth=2, label='15-min threshold')
axes[1].set_xlabel('Arrival Delay (minutes)')
axes[1].set_ylabel('Count')
axes[1].set_title('Arrival Delay Distribution (clipped to [-60, 180])')
axes[1].legend()

plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/target_distribution.png', dpi=150)
plt.close(fig)
```

## 4. Feature Distributions

```python
# Numerical feature distributions
num_features = ['CRSDepTime', 'CRSArrTime', 'CRSElapsedTime', 'Distance', 'DepDelay']

fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(num_features):
    data = df_valid[col].dropna()
    axes[i].hist(data, bins=50, color='steelblue', edgecolor='white')
    axes[i].set_title(f'{col}\n(mean={data.mean():.1f}, std={data.std():.1f})')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

axes[-1].axis('off')
plt.suptitle('Numerical Feature Distributions', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
fig.savefig(f'{PLOTS_DIR}/numerical_distributions.png', dpi=150)
plt.close(fig)
```

```python
# DepDelay vs ArrDelay scatter — visualizing the data leakage concern
fig, ax = plt.subplots(figsize=(8, 8))
sample = df_valid.sample(n=5000, random_state=42)
ax.scatter(sample['DepDelay'], sample['ArrDelay'], alpha=0.3, s=10, color='steelblue')
ax.axhline(y=15, color='red', linestyle='--', linewidth=1.5, label='ArrDelay = 15 min threshold')
ax.axvline(x=15, color='#dd8452', linestyle='--', linewidth=1.5, label='DepDelay = 15 min')
ax.set_xlabel('Departure Delay (min)')
ax.set_ylabel('Arrival Delay (min)')
ax.set_title(f'DepDelay vs ArrDelay (r = {df_valid["DepDelay"].corr(df_valid["ArrDelay"]):.3f})')
ax.set_xlim(-30, 200)
ax.set_ylim(-60, 200)
ax.legend()
plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/depdelay_vs_arrdelay.png', dpi=150)
plt.close(fig)
```

```python
# Feature distributions by delay status — DepDelay and Distance
on_time = df_valid[df_valid['Delayed'] == 0]
delayed = df_valid[df_valid['Delayed'] == 1]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# DepDelay by delay status
axes[0].hist(on_time['DepDelay'].clip(-30, 120), bins=60, alpha=0.6, color='#4c72b0', label='On-Time', edgecolor='white')
axes[0].hist(delayed['DepDelay'].clip(-30, 120), bins=60, alpha=0.6, color='#dd8452', label='Delayed', edgecolor='white')
axes[0].set_xlabel('Departure Delay (min)')
axes[0].set_ylabel('Count')
axes[0].set_title('DepDelay Distribution by Delay Status')
axes[0].legend()

# Distance by delay status
axes[1].hist(on_time['Distance'], bins=50, alpha=0.6, color='#4c72b0', label='On-Time', edgecolor='white')
axes[1].hist(delayed['Distance'], bins=50, alpha=0.6, color='#dd8452', label='Delayed', edgecolor='white')
axes[1].set_xlabel('Distance (miles)')
axes[1].set_ylabel('Count')
axes[1].set_title('Distance Distribution by Delay Status')
axes[1].legend()

plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/features_by_delay_status.png', dpi=150)
plt.close(fig)
```

```python
# Flights by Day of Week (only 1 month of data, so month chart is not useful)
day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
day_counts = df_valid['DayOfWeek'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(day_labels, day_counts.values, color='steelblue')
ax.set_title('Flights by Day of Week')
ax.set_ylabel('Count')
plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/flights_by_day.png', dpi=150)
plt.close(fig)
```

## 5. Correlation Analysis

```python
# Correlation heatmap of key numerical features
corr_cols = ['DepDelay', 'ArrDelay', 'CRSDepTime', 'CRSArrTime', 'CRSElapsedTime',
             'Distance', 'DayOfWeek', 'DayofMonth', 'TaxiOut', 'TaxiIn', 'AirTime']
corr_cols = [c for c in corr_cols if c in df_valid.columns]

corr_matrix = df_valid[corr_cols].corr()

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5, ax=ax)
ax.set_title('Correlation Matrix — Key Numerical Features')
plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/correlation_heatmap.png', dpi=150)
plt.close(fig)

print(f"DepDelay ↔ ArrDelay correlation: {df_valid['DepDelay'].corr(df_valid['ArrDelay']):.3f}")
print('→ Very high correlation confirms potential data leakage concern from proposal.')
```

## 6. Temporal Patterns

```python
# Delay rate by day of week and hour of day (month excluded — dataset only has January)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

# By Day of Week
delay_by_day = df_valid.groupby('DayOfWeek')['Delayed'].mean() * 100
axes[0].bar(day_labels, delay_by_day.values, color='#dd8452')
axes[0].set_title('Delay Rate by Day of Week')
axes[0].set_ylabel('Delay Rate (%)')
axes[0].set_ylim(0, delay_by_day.max() * 1.3)
for i, v in enumerate(delay_by_day.values):
    axes[0].text(i, v + 0.5, f'{v:.1f}%', ha='center', fontsize=10)

# By Hour of Day
df_valid['DepHour'] = df_valid['CRSDepTime'] // 100
delay_by_hour = df_valid.groupby('DepHour')['Delayed'].mean() * 100
axes[1].plot(delay_by_hour.index, delay_by_hour.values, color='#dd8452', marker='o', linewidth=2)
axes[1].set_title('Delay Rate by Scheduled Departure Hour')
axes[1].set_xlabel('Hour of Day')
axes[1].set_ylabel('Delay Rate (%)')
axes[1].set_xticks(range(0, 24))
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/delay_rate_temporal.png', dpi=150)
plt.close(fig)
```

## 7. Route & Airport Analysis

```python
# Top 15 busiest origin and destination airports
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

top_origins = df_valid['Origin'].value_counts().head(15)
top_origins.plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('Top 15 Origin Airports by Flight Count')
axes[0].set_xlabel('Number of Flights')
axes[0].invert_yaxis()

top_dests = df_valid['Dest'].value_counts().head(15)
top_dests.plot(kind='barh', ax=axes[1], color='steelblue')
axes[1].set_title('Top 15 Destination Airports by Flight Count')
axes[1].set_xlabel('Number of Flights')
axes[1].invert_yaxis()

plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/top_airports.png', dpi=150)
plt.close(fig)
```

```python
# Airports with highest delay rates (minimum 100 flights)
origin_delay = df_valid.groupby('Origin').agg(
    flights=('Delayed', 'count'),
    delay_rate=('Delayed', 'mean')
).query('flights >= 100').sort_values('delay_rate', ascending=False)
origin_delay['delay_rate'] *= 100

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

top_delayed = origin_delay.head(15)
axes[0].barh(top_delayed.index, top_delayed['delay_rate'], color='#dd8452')
axes[0].set_title('Top 15 Most Delayed Origin Airports (min 100 flights)')
axes[0].set_xlabel('Delay Rate (%)')
axes[0].invert_yaxis()
for i, (v, n) in enumerate(zip(top_delayed['delay_rate'], top_delayed['flights'])):
    axes[0].text(v + 0.3, i, f'{v:.1f}% (n={n:,})', va='center', fontsize=9)

least_delayed = origin_delay.tail(15)
axes[1].barh(least_delayed.index, least_delayed['delay_rate'], color='#4c72b0')
axes[1].set_title('Top 15 Most Punctual Origin Airports (min 100 flights)')
axes[1].set_xlabel('Delay Rate (%)')
axes[1].invert_yaxis()
for i, (v, n) in enumerate(zip(least_delayed['delay_rate'], least_delayed['flights'])):
    axes[1].text(v + 0.3, i, f'{v:.1f}% (n={n:,})', va='center', fontsize=9)

plt.tight_layout()
fig.savefig(f'{PLOTS_DIR}/airport_delay_rates.png', dpi=150)
plt.close(fig)
```

## 8. Key Takeaways

1. **Class Imbalance** — 75.9% on-time vs 24.1% delayed (3.1:1 ratio). Will need `class_weight='balanced'` or SMOTE.

2. **Data Leakage** — `DepDelay` ↔ `ArrDelay` correlation is 0.98. Models must be trained both with and without `DepDelay`.

3. **Droppable Columns** — 50+ columns are ≥95% null (diversion fields, codeshare metadata, `Unnamed: 119`). Safe to drop.

4. **Single-Month Dataset** — The SkyWest subset only contains January 2024 data, so month and year are not useful features.

5. **Temporal Patterns** — Delay rates increase throughout the day (cascading delays), peaking in evenings. Friday has the highest delay rate by day of week.

6. **Airport Effects** — Clear variation in delay rates across airports (7.7% at BFL vs 46.5% at MBS), making `Origin` and `Dest` useful features.

7. **Feature Selection** — Key candidates: `DayOfWeek`, `DayofMonth`, `CRSDepTime`, `CRSElapsedTime`, `Distance`, `Origin`, `Dest`, and optionally `DepDelay`.
